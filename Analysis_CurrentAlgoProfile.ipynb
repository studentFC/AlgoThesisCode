# Analysis_CurrentAlgoProfile
## Block 1: Importing libraries

```python
# Packages / libraries
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

## Block 2: Data loading

```python
# Load the data
raw_data = pd.read_csv('C:\\Users\\filip\\Documents\\HSG\\Master\\Master Thesis\\Research\\Data for Python\\Data Analysis CurrentAlgoProfile.csv')

# Run the first 15 rows
raw_data.head(15)
```

## Block 3: Data visualization

```python
# Create a scatter plot
plt.scatter(raw_data['algorithmization'], raw_data['algo_profile'])
plt.xlabel('Degree of algorithmization')
plt.ylabel('Current algo profile')
plt.title('Scatter plot of degree of algorithmization vs. current algo profile')

# Set the x-axis and y-axis limits
plt.xlim(-0.5, 10.5)
plt.ylim(-0.05, 1.05)
 
# Set the tick locations and labels for the x-axis and y-axis
plt.xticks(range(11))      # range(11) function generates a list of natural numbers from 0 to 10

# Show only the gridlines of natural numbers
plt.grid(axis='both', which='major', color='grey', linestyle='-', alpha=0.2)

plt.show()
```

## Bock 4: Data cleaning

```python
# Create a copy
data = raw_data.copy(deep=True)

# Drop the index column
data = data.drop(['index'], axis = 1)

# Run the first 15 rows
data.head(15)
```

## Block 5: Model training and model evaluation

```python
# Add a constant term to the features
data_with_constant = sm.add_constant(data['algorithmization'])

# Define the target variable
target_variable = data['algo_profile']

# Fit the linear regression model
model = sm.OLS(target_variable, data_with_constant).fit()

# Print the regression summary
print(model.summary())
```
